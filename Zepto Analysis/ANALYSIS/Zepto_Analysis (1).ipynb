```python
%%sql
drop table if exists zproducts;

CREATE TABLE zproducts (
    id INT AUTO_INCREMENT PRIMARY KEY,
    Category VARCHAR(100),
    name VARCHAR(255),
    mrp DECIMAL(10,2),
    discountPercent INT,
    availableQuantity INT,
    discountedSellingPrice DECIMAL(10,2),
    weightInGms INT,
    outOfStock TINYINT(1),
    quantity INT
);
```

```python
import pandas as pd
from sqlalchemy import create_engine

df = pd.read_excel(r"D:\zepto.xlsx", sheet_name="zepto1")  


# 2. Connect to MySQL
engine = create_engine("mysql+mysqlconnector://root:password@localhost:3306/dbname")

# 3. Insert into Products table
df.to_sql("zproducts", con=engine, if_exists="append", index=False)

print("Data inserted successfully!")

```

 ### Q1.Find the total number of unique product categories.

```python
%%sql
select distinct category as unique_products from zproducts;
```

### Q2.	List the top 5 most expensive products by MRP.

```python
%%sql
select  name,max(mrp) as price from zproducts group by name order by price desc limit 5;
```

### Q3.	Get the average discount percentage for each category.

```python
%%sql
select category,avg(discountpercent)as average from zproducts group by category;
```

### Q4.	Find products that are currently out of stock.

```python
%%sql
select name from zproducts where outOfStock = True order by name ;
```

### Q5.	Show the total available stock (sum of availableQuantity) per category.

```python
%%sql
select category,sum(quantity)as Total from zproducts group by category order by Total;
```

### Q6.	Find the product with the highest discounted selling price in each category.

```python
%%sql
select name as `Name of the product`,max(discountedSellingPrice) as `DiscountedSellingPrice` from zproducts group by name order by`DiscountedSellingPrice`desc limit 1;
```

### Q7.	List products where the discount is more than 20%.

```python
%%sql
select name as `Name of the product`,discountPercent from zproducts where discountPercent > 20;
```

### Q8.	Calculate potential revenue if all available stock is sold at discountedSellingPrice.

```python
%%sql
select sum(discountedSellingPrice  * quantity) as Potential_revenue from zproducts;
```

### Q9.	Find products where discountedSellingPrice is more than 90% of MRP (low discount).

```python
%%sql
select name as `Name of the Product`, mrp,
    discountedSellingPrice from zproducts where discountedSellingPrice > 0.9 * mrp;
```

### Q10.	Show the category with the highest number of products.

```python
%%sql
select category,count( distinct name) as Total from zproducts group by category order by Total desc limit 1;
```

### Q11.Find products that weigh less than 100 grams but have high selling prices.

```python
%%sql
select name as `Name of the Product`,weightInGms,discountedSellingPrice from zproducts where weightInGms < 100
and   discountedSellingPrice > 1000 order by weightInGms,discountedSellingPrice;
```

### Q12.Calculate the average weight per category.

```python
%%sql
select category,avg(weightInGms) as AvgweightInGms from zproducts group by category order by AvgweightInGms ;
```

### Q13.	List products where availableQuantity is less than 5 (low stock alert).

```python
%%sql
select id,name as `Name of the Products`,quantity from zproducts where quantity < 5;
```

### Q14.Find the top 3 categories contributing the most revenue (availableQuantity × discountedSellingPrice).

```python
%%sql
select category,sum(availableQuantity * discountedSellingPrice) Total from zproducts
         group by category
                     order by   Total desc limit 3;
```

### Q15.Identify products where discountPercent is high but MRP is still below the overall average MRP.

```python
%%sql
SELECT name AS `Product Name`,
    mrp,
    discountedSellingPrice,
    ROUND(((mrp - discountedSellingPrice) / mrp) * 100, 2) AS discountPercent FROM zproducts
WHERE ((mrp - discountedSellingPrice) / mrp) * 100 > 30  
  AND mrp < (SELECT AVG(mrp) FROM zproducts);            
```
